```python
# Import dependencies 
import pandas as pd
```

```python
# Import csv
snow_data_to_load = "data/data_1993_2018.csv"
```

```python
# Read csv
snow_data = pd.read_csv(snow_data_to_load)
snow_data.head()
```

```python
# Drop unneeded columns
filtered_snow_data = snow_data.drop(columns=['SiteNumber', 'SampleDate', 'SampleType', 'ANC (ueq/L)', 'SC (uS/cm)', 'Total N (mg/L)', 'Total Dissolved N (mg/L)', 'Hg (ng/L) Roth 1994', 'Hg (ng/L) USEPA 1631', 'd34S (per mil)', 'SWE (cm)', 'Ionic Balance (%)'])
filtered_snow_data
```

```python
# Obtain data entry count of each Station Name
station_count = filtered_snow_data['StationName'].value_counts()
station_count = station_count[station_count >= 20]
station_count
```

```python
station_count = pd.DataFrame(station_count)
station_count.reset_index(inplace=True)
station_count.rename(columns={"index": "StationName", "StationName": "Station_Count"}, inplace=True)
station_count.head()
```

```python
# Merge dataframes
filtered_snow_data = filtered_snow_data.merge(station_count, on = 'StationName', how = 'left')
```

```python
filtered_snow_data.dropna(inplace=True)
```

```python
filtered_snow_data.head()
```

```python
filtered_snow_data.rename(columns={"H (ueq/L)": "Hydrogen", "Ca (ueq/L)": "Calcium", "Mg (ueq/L)": "Magnesium", 
          "Na (ueq/L)": "Sodium", "K (ueq/L)": "Potassium", "NH4 (ueq/L)": "Ammonium", 
          "Cl (ueq/L)": "Chloride", "SO4 (ueq/L)": "Sulfate", "NO3 (ueq/L)": "Nitrate", 
          "DOC (mg/L)": "Dissolved_organic_carbon", "Snow Depth (cm)": "Snow_depth"}, inplace=True)
```

```python
final_snow_data = filtered_snow_data.drop(columns=['Station_Count'])
```
